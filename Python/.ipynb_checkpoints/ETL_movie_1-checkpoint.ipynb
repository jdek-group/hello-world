```python
# WRONG JOIN . REFER BELOW FOR THE CORRECT JOIN
movie_and_rating_left_join = pd.merge(movie, rating, on='movieId', how='left')
# pd.set_option('display.max_rows', None)
movie_and_rating_left_join
```

```python
movie_and_rating_left_join.dtypes
```

```python
movie_and_rating_right_join = pd.merge(movie, rating, on='movieId', how='right')
movie_and_rating_right_join
```

```python
movie_and_rating_right_join.dtypes
```

```python
# checking for missing values 
has_missing_values = movie_and_rating_right_join['userId'].isnull().sum()
has_missing_values
```

```python
# Exploration
# check how many distinct users in userId
num_distinct_users = movie_and_rating_right_join['userId'].nunique()
print(f'Number of distinct users: {num_distinct_users}')
```

```python
# Step 1: Extract unique userId values and sort them
unique_user_ids = movie_and_rating_right_join['userId'].dropna().unique()

# Step 2: Select the first 10,000 users
# selected_users = unique_user_ids[:10000] #out of memory 150k+ rows
# selected_users = unique_user_ids[:500]   #out of memory (71554, 6)
selected_users = unique_user_ids[:100]     #(11101, 6)

# Step 3: Filter the DataFrame to include only rows with the selected userIds
sample_10000_df = movie_and_rating_right_join[movie_and_rating_right_join['userId'].isin(selected_users)]

# View the resulting DataFrame
# sample_10000_df.head(1000)
sample_10000_df
```
